```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "gemma-3-1b"
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from datasets import load_dataset

# Replace 'path_to_your_dataset' with the actual path or dataset name
dataset = load_dataset('"C:\Users\Owner\Downloads\tested.csv"')

def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True, padding='max_length', max_length=512)

tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=3,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],
    eval_dataset=tokenized_dataset['validation'],
)
```

```python
trainer.train()
```

```python
from transformers import TrainerCallback

class TensorBoardCallback(TrainerCallback):
    def on_log(self, args, state, control, logs=None, **kwargs):
        import tensorboardX as tb
        writer = tb.SummaryWriter(log_dir=args.logging_dir)
        for k, v in logs.items():
            writer.add_scalar(k, v, state.global_step)

trainer.add_callback(TensorBoardCallback())
```
